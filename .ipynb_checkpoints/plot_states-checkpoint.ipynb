```python
import os
import pandas as pd
import plotly.graph_objects as go
```

## Plot R_t on a map

```python
# Read in Rt data
rt_df = pd.read_csv('https://d14wlfuexuxgcm.cloudfront.net/covid/rt.csv')
```

```python
dates = rt_df['date'].unique()
states = rt_df['state'].unique()
```

```python
# Add missing dates with dummy values (0)
for date in dates:
    states_covered = set(rt_df[rt_df['date'] == date]['state'].unique())
    states_not_covered = set(states).difference(states_covered)
    for state in states_not_covered:
        new_df = pd.DataFrame(
            data={
                'state': [state],
                'date': [date],
                'ML': [0],
                'Low_90': [0],
                'High_90': [0],
                'Low_50': [0],
                'High_50': [0]
            }
        )
        rt_df = rt_df.append(new_df, ignore_index=True)    
```

```python
# Set min and max for the color bar
zmin, zmax = rt_df['ML'].min(), rt_df['ML'].max()
```

```python
# Create Choropleth plot
for date in dates:
    date_df = rt_df[rt_df['date']==date]
    
    fig = go.Figure(data=go.Choropleth(
        locations=date_df.index,
        z = date_df['ML'],
        locationmode='USA-states',
        colorscale='Reds',
        colorbar_title='R_t',
        marker_line_color='white',
        zauto=False,
        zmin=zmin,
        zmax=zmax,
    ))

    fig.update_layout(
        title_text = 'Effective reproductive number (R_t) of COVID-19 by state',
        annotations=[
            dict(
                x=0.5,
                y=-0.15,
                showarrow=False,
                text=date,
                font=dict(
                    family='Arial, sans-serif',
                    size=18,
                    color='black'
                ),
            ),
        ],
        geo = dict(
            scope='usa',
            projection=go.layout.geo.Projection(type='albers usa'),
            showlakes=True,
            lakecolor='rgb(255, 255, 255)'),
        )

    fig.write_image('plots/{}.jpg'.format(date))
```
